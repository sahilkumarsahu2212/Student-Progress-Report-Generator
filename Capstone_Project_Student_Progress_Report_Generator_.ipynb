```python
print("="*80)
print("Generate a Consolidated progress report for each student by merging data from:")
print("="*80)
print("\t\t\tStudent Details (csv file)")
print("\t\t\tMarks Table (csv file)")
print("\t\t\tAttendence Record (csv file)")
print("="*80)
print("Problem Requirements")
print("\t1.Combine all three datasets")
print("\t2.For each student calculate")
print("\t\t *.Total Obtained Mark")
print("\t\t *.Total Max Mark")
print("\t\t *.Percentage")
print("\t\t *.Attendance Percentage")
print("\t3.Generate Final Status")
print("="*80)
import pandas as pd

#Load data from CSV file
students=pd.read_csv("students.csv")
marks=pd.read_csv("marks.csv")
attendance=pd.read_csv("attendence.csv")

#stpe-1 combine all three datasets(aggregate marks)
marks_summary=marks.groupby("roll_no").agg(
    total_obtained_mark=("marks_obtained","sum"),
    total_max_mark=("max_marks","sum")
).reset_index()
print(marks_summary)

#step-2  join students+marks
report=pd.merge(students,marks_summary,on="roll_no",how="left")

#stpe-3   join with attendence, total we joined students+marks+attendence tables
report=pd.merge(report,attendance,on="roll_no",how="left")

#step-4   Calculate percentage and attendance percentage
report['marks_percentage']=report['total_obtained_mark']/report['total_max_mark']*100
report['attendance_percentage']=report['present_days']/report['working_days']*100

#step-5   Final Results
report["status"]=report.apply(
    lambda row: "Pass" if row["marks_percentage"] >=60 and row["attendance_percentage"] >=75 else "Fail",
    axis=1
)
print(report)
```
